```python
raw = pd.read_excel('speeches.xlsx')
raw.drop('info', inplace=True, axis=1)
raw[25:30]
```

```python
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
nltk.download('stopwords')

# lower case for all the reviews
def lowerize(x):
  list1 = []
  for i in x:
    list1.append(str.lower(i))
  return list1


freq = raw.copy()

# lower case the whole speech
freq['speech'] = freq['speech'].str.lower()

# creating a column without stop words
stop = stopwords.words('english')
freq['no_stopwords'] = freq['speech'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

# Removing punctuation from no_stopwords
freq['no_stopwords'] = freq['no_stopwords'].str.replace(r'[^\w\s]+', '')

# tokenizing
freq['tokenized'] = freq['no_stopwords'].apply(word_tokenize)
freq
```

```python
df_trump_1 = freq[freq['name']=='Donald Trump'] #a dataframe to store only Trump speeches

#find the frequency of each non-stopword for Trump speeches
unsorted_frequency_trump = df_trump_1.tokenized.str.get_dummies(',').sum()
sorted_frequency_trump = unsorted_frequency_trump.sort_values(ascending = False)
```

```python
# data frame of attributes
columns = ['frequency']
word_freq_trump = pd.DataFrame(sorted_frequency_trump[:60], columns = columns)
word_freq_trump
```

Now we will compute Lifts for them with 15 previous Presidents (excluding Trump)



```python
list_of_attributes=['people', 'us', 'you', 'great', 'one', 'american', 'united']
lift_matrix_attributes=pd.DataFrame(columns=list_of_attributes)

former_presidents_15 = list(freq['name'].unique())[:15]



lift_matrix_attributes['President']=former_presidents_15
lift_matrix_attributes=lift_matrix_attributes.set_index('President')
#created a square matrix to store brand-and-feature list values
```

```python
from math import sqrt

size_lift_matrix_features_rows=int(len(former_presidents_15)) #storing no. of rows of lift matrix (for iteration purpose)
size_lift_matrix_features_columns=int(len(list_of_attributes)) #storing no. of rows of lift matrix (for iteration purpose)

def tolistfunc(X):
  A=[]
  A.append(X)
  return A


freq['list_president'] = freq['name'].map(tolistfunc)
freq['tokenized_president'] = freq['tokenized'] + freq['list_president']
```

```python
N=len(freq)

i=0
j=0
for i in range(0,size_lift_matrix_features_rows):
    for j in range(0,size_lift_matrix_features_columns):
        Na=0
        Nb=0
        Nab=0
        for k in range(0,len(freq)):
            if former_presidents_15[i] in freq['tokenized_president'][k]:
                  Na=Na+1
            if list_of_attributes[j] in freq['tokenized_president'][k]:
                  Nb=Nb+1
            if (former_presidents_15[i] in freq['tokenized_president'][k]) and (list_of_attributes[j] in freq['tokenized_president'][k]):
                  Nab=Nab+1       
        lift_matrix_attributes.iat[i,j]=(N*Nab)/(Na*Nb) 
```
